Pi to the nth digit

```python
import math
def pi_to_nth(n):
    num = math.pi
    factor = 10 ** n
    num = int(num * factor)
    num = float(num / factor)
    return num
```

Fibonacci Sequence

```python
def fibonacci_sequence(n):
    my_list = [0,1]
    next_num = 1
    while next_num < n:
        my_list.append(next_num)
        next_num = my_list[-1] + my_list[-2]
    return my_list
```

```python
print(fibonacci_sequence(25))
```

Collatz Conjecture

```python
def collatz_conjecture(n):
    step_count = 0
    while n > 1:
        if n % 2 == 0:
            n = int(n / 2)
            step_count += 1
        else:
            n = (n * 3) + 1
            step_count += 1
    print(f'Step Count: {step_count}')
```

Sieve of Eratosthenes

```python
import math

def sieve_of_eratosthenes(n):
    numbers = [num for num in range(2,n+1)]
    composite = set()
    p = 2
    while math.pow(p, 2) <= n:
        for i in numbers:
            if i % p == 0 and i != p:
                composite.add(i)
        p += 1
    prime_numbers = [num for num in numbers if num not in composite]
    print(prime_numbers)
```

```python
sieve_of_eratosthenes(20)
```

```python
def pig_latin_multi(words):
    
    word_list = [word for word in words.split(' ')]
    
    for word in word_list:
        pig_latin(word)

def pig_latin(word):
    
    valid_word = check_word_validity(word)
    
    if not valid_word:
        print('Invalid Word')
        return
    
    constanant_position = find_constanant(word)
    sub_string = f'{word[constanant_position]}ay'
    
    if word[0] == 'y':
        pig_latin_word = word[1::] + 'yay'
    elif starts_with_constanant(word):
        cluster = get_constanant_cluster(word)
        pig_latin_word = word[len(cluster)::] + f'-{cluster}ay'
    else:
        pig_latin_word = word + 'yay'
    print(pig_latin_word)
        

def check_word_validity(word):
    
    for c in word:
        if not c.isalpha():
            return False
        
    return True
        
        
def find_constanant(word):
    
    vowels = ['a','e','i','o','u']
    
    for i in range(len(word)):
        if word[i] not in vowels:
            return i
        
def starts_with_constanant(word):

    vowels = ['a','e','i','o','u']
    
    if word[0] in vowels:
        return False
    else:
        return True
    
def get_constanant_cluster(word):
    
    cluster = ''
    
    for c in word:
        is_vowel = check_vowel(c)
        if is_vowel:
            return cluster
        else:
            cluster = cluster + c 
        
def check_vowel(char):
    vowels = ['a','e','i','o','u']
    if char in vowels:
        return True
    else:
        return False
```

Count Vowels

```python
def count_vowels(word):
    
    vowels = ['a','e','i','o','u']
    vowel_count = 0
    
    for c in word:
        if c in vowels:
            vowel_count += 1
    
    return vowel_count
```

Check if Palindrome

```python
def check_palindrome(word):
    
    reverse = word[::-1]
    
    if word == reverse:
        print(f'{word} is a palindrome')
        return
    else:
        print(f'{word} is not a palindrome')
        return
```

<h4>Project Inventory Project</h4>
<p>Create an application which manages an inventory of products. Create a product class which has a price, id, and quantity on hand. Then create an inventory class which keeps track of various products and can sum up the inventory value.</p>

```python
class product():
    
    
    def __init__(self,product_id,price,quantity):
        self.id = product_id
        self.price = price
        self.quantity = quantity
        
    def __len__(self):
        return self.quantity
        
    def stock_added(self, quantity_added):
        self.quantity += quantity_added
        
    def stock_removed(self, quantity_removed):
        self.quantity -= quantity_removed
        
    def product_value(self):
        return self.price * self.quantity
        
class inventory():
    
    def __init__(self, products):
        self.products = [product for product in products]
        
    def __len__(self):
        return len(self.products)
        
    def sum_inventory_value(self):
        
        value = 0
        
        for product in self.products:
            product_value = product.product_value()
            value += product_value
            
        return value
```

<p>Create an hierarchy of classes - abstract class Employee and subclasses HourlyEmployee, SalariedEmployee, Manager and Executive. Every one's pay is calculated differently, research a bit about it. After you've established an employee hierarchy, create a Company class that allows you to manage the employees. You should be able to hire, fire and raise employees.</p>

```python
class employee():
    
    def __init__(self):
        print('employee created')
        
    def calculate_pay(self):
        pass
```

```python
class hourly_employee(employee):
    
    def __init__(self, wages, weekly_hours):
        employee.__init__(self)
        self.wages = wages
        self.hours = weekly_hours
        print(f'hourly employee created on £{self.wages} an hour')
        
    def calculate_pay(self):
        weekly_pay = self.wages * self.hours
        print(f'hourly employee weekly pay: £{weekly_pay}')
        annual_pay = weekly_pay * 52
        print(f'hourly employee annual pay: £{annual_pay}')
        return annual_pay
```

```python
class salaried_employee(employee):
    
    def __init__(self, salary):
        employee.__init__(self)
        self.salary = salary
        print(f'salaried employee created on £{self.salary} per year')
        
    def calculate_pay(self):
        print(f'salaried employee annual pay: £{self.salary}')
        return self.salary
```

```python
class manager(salaried_employee):
    
    def __init__(self, salary):
        salaried_employee.__init__(self, salary)
        print('manager created')
        
    def calculate_pay(self):
        manager_pay = salaried_employee.calculate_pay(self)
        print(f'manager annual pay: £{manager_pay}')
        return manager_pay
```

```python
class executive(salaried_employee):
    
    def __init__(self, salary):
        employee.__init__(self, salary)
        print('executive created')
        
    def calculate_pay(self):
        executive_pay = salaried_employee.calculate_pay(self)
        print(f'executive annual pay: £{executive_pay}')
        return executive_pay
```
